```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from skimage.measure import label, regionprops
from skimage.segmentation import clear_border
from skimage.color import label2rgb
```

### Load classification data

```python
fname = 'counting_example.npz'
data = np.load(fname)['arr_0']
data = np.squeeze(data)
```

```python
fig = plt.figure()
fig.add_subplot(1, 2, 1)
plt.imshow(data)
fig.add_subplot(1, 2, 2)
out = plt.hist(data.flatten(), bins = 10)
```

## Threshold the binary image

```python
threshold = 0.7
data[data > threshold] = 1
data[data <= threshold] = 0
plt.imshow(data)
```

## Count blobs

```python
label_image = label(data)
image_label_overlay = label2rgb(label_image, image=data)
```

```python

fig = plt.figure(figsize=(10, 10))
fig.add_subplot(1, 2, 1)
plt.imshow(data)
fig.add_subplot(1, 2, 2)
plt.imshow(image_label_overlay)
count = np.unique(label_image)[-1]
plt.title('Count: ' + str(count))
```
