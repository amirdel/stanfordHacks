```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import os
```

List images

```python
train = [os.path.join('/root/data/hackathon/building_massa_dataset/images/train/', f ) for f in os.listdir('/root/data/hackathon/building_massa_dataset/images/train/') if f.endswith('.tiff')]
```

```python
img_list = np.random.choice(train, size = 10)
```

```python
fig, ax = plt.subplots(10, 1, figsize=(10, 40))

for (i, img) in enumerate(img_list):
    # create the truth from the image path
    ftruth = img.replace('images', 'labels')
    ftruth = ftruth[:-1]
    ax[i].imshow(np.array(Image.open(img)))
    ax[i].imshow(np.array(Image.open(ftruth)), alpha=0.5)
```
